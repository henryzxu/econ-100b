# Section 5 Graphs

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k_over_l_values = np.arange(0, 200001, 1)
alpha = [0.25, 0.5, 0.75, 0.99]
E = np.arange(10000, 50000, 10000)
```

```python
for a in alpha:
    y_over_l_values = k_over_l_values ** a * E[0] ** (1-a)
    plt.plot(k_over_l_values, y_over_l_values, label = "alpha = %0.2f" % a)
plt.legend()
plt.xlabel('K/L')
plt.ylabel('Y/L')
plt.title('Cobb-Douglas Production Function with Varying Alpha')
plt.show()
```

```python
for e in E:
    y_over_l_values = k_over_l_values ** alpha[0] * e ** (1-alpha[0])
    plt.plot(k_over_l_values, y_over_l_values, label = "E = %d" % e)
plt.legend()
plt.xlabel('K/L')
plt.ylabel('Y/L')
plt.title('Cobb-Douglas Production Function with Varying E')
plt.show()
```
